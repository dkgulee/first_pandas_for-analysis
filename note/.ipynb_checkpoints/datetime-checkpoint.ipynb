# 시계열 데이터
### 시계열 데이터는 데이터 분석 분야에서 중요하게 다루는 데이터 중 하나다. 우리가 지금까지 다룬 날씨 관측 데이터, 에볼라 전염병으로 인한 사망자 수 관측 데이터, 빌보드 차트 데이터에는 모두 시계열 데이터가 포함되어 있다. 즉 일정 시간 간격으로 어떤 값을 기록한 데이터에서는 시계열 데이터가 매우 중요하다! 그렇기 떄문에 시계열 데이터를 자유자재로 다룰 줄 알아야 한다


# datetime 오브젝트
### datetime 라이브러리는 날짜와 시간을 처리하는 등 다양한 기능을 제공하는 파이선 라이브러리다. <br>날자를 처리하는 date 시간처리 time 날짜시간 모두 datetime

```python
from datetime import datetime
```

```python
now1 = datetime.now()
print(now1)
```

```python
now2 =datetime.today()
print(now2)
print(type(now2))
```

## datetime()에 인자값을 넘겨  datetime 오브잭트가 생김

```python
t1 = datetime.now()
t2 = datetime(1970, 1, 1)
t3 = datetime(1970, 12, 12, 13, 24, 34)
print(t1)
print(t2)
print(t3)
```

## datetime 오브젝틑 사용하는 이유 중 하나는 시간 계산을 할 수 있다!

```python
diff1 = t1 - t2
print(diff1)
dfii2 = t2 - t1
print(dfii2)
```

# dt 접근자 사용하기

### 문자열을 처리하려면 str 접근자를 사용한 다음 문자열 속성이나 메서드를 사용해야 했다.<br> datetime 오브젝트도 마찬가지로 dt 접근자를 사용하면 datetime 속성이나 메서드를 사용하여 시계열 데이터를 처리할 수 있다!

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
ebloa.head()
```

```python
ebola['date_dt'] =pd.to_datetime(ebola['Date'])
ebola
```

```python
ebola[['Date', 'date_dt']].head()
```

## dt 접근자를 사용하지 않고 인덱스가 3인 데이터 년월 일 데이터 추출

```python
print(ebola['date_dt'][3].year)
print(ebola['date_dt'][3].month)
print(ebola['date_dt'][3].day)
```

```python
ebola['year'] = ebola['date_dt'].dt.year
ebola[['Date', 'date_dt', 'year']]
```

```python
ebola['month'] = ebola['date_dt'].dt.month
ebola[['Date', 'date_dt', 'year', 'month']]
ebola['day'] = ebola['date_dt'].dt.day
ebola[['Date', 'date_dt', 'year', 'month', 'day']]
```

```python
ebola.info()
```

# 사례별 시계열 데이터 계산하기

```python
print(ebola.iloc[-5:, :5])
print(ebola['date_dt'].min())
print(type(ebola['date_dt'].min()))
```

## 에볼라 진행 전동 알아보기 발생 시간 - 발생년도 최소값

```python
ebola['outbreak_day'] = ebola['date_dt'] - ebola['date_dt'].min()
```

```python
print(ebola[['Date', 'Day', 'outbreak_day']].head())
```
