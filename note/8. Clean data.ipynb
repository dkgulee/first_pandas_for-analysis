# 열이름 관리하기
### 어떤 열은 여러가지 의미를 가지고 있을 수 있다.
### 예를 들어 ebola 의 Deaths_Guinea 는 '사망자 수' 와 '나라 이름'을 합쳐 만든 이름이다

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
print(ebola.columns)
print(ebola.iloc[:5,[0, 1, 2, 3, 10, 11]])
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
print(ebola_long.head())
```

## split 메서드로 열 이름 분리하기
### Gases-Guinea 와 같이 2개 이상의 의미를 가지고 있는 열이다.
### split을 사용해서 Gases, Guinea 와 같은 방법으로 분리 할 수 있다.

###  Series , 한개는 list  

```python
variabel_split = ebola_long.variable.str.split('_')
print(variabel_split[:5])
```

```python
print(type(variabel_split[:5]))
print(type(variabel_split[0]))
```

```python
status_values = variabel_split.str.get(0)
country_values = variabel_split.str.get(1)

print(status_values[:5])
```

```python
print(status_values[-5:])
```

```python
print(country_values[:5])
print(country_values[-5:])
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
print(ebola_long.head())
print(status_values.head())
print(country_values.head())
```

# concat 메서드로 데이터프레임에 열 추가하기
## concat 메서드를 활용해서  split으로 분리한 데이터를 바로 데이터 프레임에 추가하기

```python
variable_split = ebola_long.variable.str.split('_', expand = True)
variable_split.columns = ['status','country']
ebola_parsed = pd.concat([ebola_long, variable_split], axis= 1)
print(ebola_parsed.head())
```

# 여러 열을 하나로 정리하기
보통 데이터프레임의 열은 파이썬의 변수와 같은 개념으로 사용합니다. 하지만 비슷한 성질의 데이터를 관리하기 위해 열이 여러 개로 분리되어 있는 경우도 있다.

### 가상의 데이터의 여러 열을 하나로 정리하기 - melt, pivot, table 메서드


```python
weather = pd.read_csv('../data/weather.csv')
print(weather.iloc[:5, :])
```

## melt 매소드 활용하기

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day',
                      value_name ='temp')
print(weather_melt.head())
```

### pivot_table 메서드는 행과 열의 위치를 다시 바꿔 정리해준다.

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns = 'element',
    values = 'temp',
    dropna=False
)
print(weather_tidy.head())
```

```python
weather_tidy_flat = weather_tidy.reset_index()
print(weather_tidy_flat.head())
```

# 중복 데이터 처리하기


```python
billboard = pd.read_csv('../data/billboard.csv')
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
                        var_name ='week', value_name ='rating')
billboard.head()
```

```python
billboard_long.head()
billboard_long.shape
```

###  노래 제목이 loser 인 데이터만 따로 모아논 곡만이 있다 (중복데이터)

```python
billboard_long[billboard_long.track == 'Loser']
```

### 중복 데이터를 가지고 있는 열  'year', 'artist', 'track', 'time', 'date.entered' 이다. 이열을 따로 모아 새로운 데이터프레임에 저장한다.

```python
billboard_songs = billboard_long[['year', 'artist', 'track', 'time', 'date.entered']]
print(billboard_songs.shape)
```

#### drop_duplicates() 메서드로 중복 데이터를 제거

```python
billboard_songs =  billboard_songs.drop_duplicates()
print(billboard_songs.shape)
type(billboard_songs)
```

### 중복을 제거하 데이터프레임에 id 추가

```python
billboard_songs['id'] = range(1,len(billboard_songs)+1)
print(billboard_songs.head(n= 10))
```

### merge 메서드를 사용해 노래 정보와 주간 수위데이터를 합친 것

```python
billboard_ratings = billboard_long.merge(billboard_songs, on=['year', 'artist', 'track', 'time', 'date.entered'])
print(billboard_ratings.head())

print(billboard_ratings.shape)
```

# 대용량 데이터 처리하기
### 여러 개로 나누어진 데이터 불러오기
데이터를 나누어 저장하면 용량이 작아져 데이터를 저장하거나 다른 사람에게 공유할 때 유용 
여태까지는 한 덩어리의 데이터를 불러와 여러 가지 실습을 함. 그러면 여러 개로 나누어진 데이터를 신속하게 불러오려면 어떻게 해야할까요?

```python
import os
import urllib.request

with open('../data/raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        if line == 5:
            break;
        fn = url.split('/')[-1].strip()
        fp = os.path.join('', '../data', fn)
        print(url)
        print(fp)
        urllib.request.urlretrieve(url, fp)
```
