# datetime 오브젝트
### datetime 라이브러리는 날짜와 시간을 처리하는 등 다양한 기능을 제공하는 파이선 라이브러리다. <br>날자를 처리하는 date 시간처리 time 날짜시간 모두 datetime

```python
from datetime import datetime
```

```python
now1 = datetime.now()
print(now1)
```

```python
now2 =datetime.today()
print(now2)
print(type(now2))
```

## datetime()에 인자값을 넘겨  datetime 오브잭트가 생김

```python
t1 = datetime.now()
t2 = datetime(1970, 1, 1)
t3 = datetime(1970, 12, 12, 13, 24, 34)
print(t1)
print(t2)
print(t3)
```

## datetime 오브젝틑 사용하는 이유 중 하나는 시간 계산을 할 수 있다!

```python
diff1 = t1 - t2
print(diff1)
dfii2 = t2 - t1
print(dfii2)
```

# dt 접근자 사용하기

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
ebloa.head()
```
